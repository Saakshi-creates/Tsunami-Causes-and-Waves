# DATA CLEANING

```python
source.columns
```

```python
#DELETING THE COLUMNS WITH THE LEAST NUMBER OF VALUES PRESENT
#For sources and waves we will calculate the percentage of values present with respect to the total values

percent_missing_source = source.isnull().sum() * 100 / len(source)
percent_missing_source
```

```python
percent_missing_waves=100*(waves.isnull().sum())/(len(waves))
percent_missing_waves
```

```python
#Dropping the columns with more than 80% of empty rows

percent_missing_source_delete=percent_missing_source[percent_missing_source>80]
percent_missing_source_delete

percent_missing_waves_delete=percent_missing_waves[percent_missing_waves>80]
percent_missing_waves_delete
```

```python
#Creating a new dataset for source as "s" and waves as "w" so that we can look at the insights that these less than 20% values can provide us with

s=source
w=waves

s.shape
```

```python
s.drop(columns=percent_missing_source_delete.index,inplace=True)
```

```python
s.shape
```

```python
w.shape
```

```python
w.drop(columns=percent_missing_waves_delete.index,inplace=True)
```
